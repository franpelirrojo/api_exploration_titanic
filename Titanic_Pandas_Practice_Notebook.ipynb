##  Step 2: Selecting Data
Las Series y los DataFrames en pandas implementan el operador de indexaci贸n de python `[]`, a la vez permiten indexar contra sus atributos `loc`y `矛loc`.

### df["column"]
Uno de los m茅todos para acceder a las columnas es indexarlas directamente desde el objeto. Igualmente podemos acceder a ellas directamente c贸mo un atributo, pero unicamente si el nombre de la columna es un nombre v谩lido par aun atributo en python. **Muchas referencias recomiendan ignorar tanto las columnas c贸mo atributo c贸mo indexar directamente desde el objeto en favor de loc y iloc**.

### df.iloc[]
El atributo iloc nos permite extraer filas y columnas mediante su posici贸n.

```python
df.iloc[4] #Es una Serie aunque es una fila!!!
```

```python
df.iloc[[4]] #Al indexar de nuevo sobre una lista, el objeto devuelto es un DataFrame
```

Al igual que en python, podemos pasarle un "slice" o incluso una funci贸n, que reciva un DataFrame y devuelva opciones v谩lidas para 'iloc'.

```python
df.iloc[0:11:5]
```

```python
df.iloc[(lambda df:[0,5,10])]
```

Y manteniendo la l贸gica de las operaciones anteriores, podemos extraer igualmente columnas.

```python
df.iloc[:,[3]]
```

### df.loc[]
El atributo loc nos permite indexar por el nombre del indice y de la columna. Es mportante recordar en este punto que los indices, aunque sean numericos autoincrementales, **son etiquetas no posiciones**.

```python
df.loc[50:100:10, ["Name", "Sex", "Pclass"]]
```

Fijate que el comportamiento de `iloc` y `loc` al definir intervalos es diferente. `iloc` implementa una l贸gica de intervalo medio abierto, incluyendo el primer indice pero no el 煤ltimo; mientras que `loc`implementa un intervalo cerrado, incluyendo tanto el primer indice c贸mo el 煤ltimo. Hay excepciones c贸mo la indexaci贸n parcial que hace que esta l贸gica se sienta algo inconssitente.

```python
df.iloc[50:100:10] #el mismo intervalo pero con iloc, observa la diferencia.
```

Podemos hacer indexaci贸n por rango en indices de strings, incluso podemos hacer una **indexaci贸n parcial**, pero los indices deben estar *ordenados*.

```python
df.set_index("Embarked").sort_index().loc['C':'Q']
```

```python
#Muy 煤til para seleccionar largas listas de nombres.
#Aqu铆 extraemos los apellidos de la A a la C.
#Importante destacar que se la indexaci贸n parcial excluye el ultimo indice:
#los apellidos empezados por D no aparecen.
df.set_index('Name').sort_index().loc["A":"D", "PassengerId":"Pclass"]
```

Puedes hacer indexaci贸n parcial de columnas si las ordenas. 

```python
df.set_index('Name').sort_index().sort_index(axis='columns').loc["A":"B", "P":"T"].head()
```
