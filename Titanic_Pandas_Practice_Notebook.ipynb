##  Step 5: Grouping and Aggregating

### df.groupby
Agrupar o pivotar la informaci贸n permite convertir en tabla la expresi贸n "obtener X por Y valor": 
- Colocar Y en el indice.
- Una columna por cada valor de X.
- Aplicar alguna transformaci贸n o de agregaci贸n si es necesario.

Esta operaci贸n divide la informaci贸n en grupos seg煤n el criterio dado y **devuelve un objeto lazy**, que no se calcula hasta que se le aplica una operaci贸n de transformaci贸n.

![Operaci贸n de agrupaci贸n](groupoperation.png)

```python
# 1. Agrupamos por Edad y lugar de embarque.
# 2. Contamos el tipo de embarque para los valores de la tabla.
# 3. Usamos unstack para sacar el indice m谩s a la derecha.
# 4. Rellenamos los NaN con ceros por presentaci贸n.
df.groupby(["Age", "Embarked"])["Embarked"].count().unstack().fillna(0.0)
```

### df.merge
